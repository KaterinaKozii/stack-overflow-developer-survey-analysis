# 2. Аналіз повноти відповідей респондентів.

**Мета завдання:**
визначити, скільки респондентів відповіли на **всі питання**, які наявні в schema.

```python
# Хоча ми це вже знаємо, ще раз перевіримо наявність стовпця qname у schema

'qname' in df1.columns
```

```python
# Отримаємо список технічних назв питань зі schema['qname'].
# dropna() прибере пропущені значення в самій schema.
# astype(str) гарантує, що всі назви будуть рядками.
df1_questions = set(df1['qname'].dropna().astype(str))

# Отримаємо множину стовпців основного датасету.
dataset_columns = set(df.columns)

# Знаходимо перетин двох множин.
# intersection() залишає тільки ті qname, які реально існують як стовпці в df.
question_columns = sorted(df1_questions.intersection(dataset_columns))

print('Кількість питань в schema[qname]:', len(df1_questions))
print('Кількість стовпців в основному датасеті:', len(dataset_columns))
print('Кількість питань, знайдених і в schema, і в основному датасеті:', len(question_columns))

# Виводимо перші 20 знайдених стовпців для перевірки.
question_columns[:20]
```

```python
# Створюємо таблицю тільки з тими стовпцями, які є питаннями зі schema та є в основному датасеті.
answers_to_schema_questions = df[question_columns]

# notna() повертає True, якщо значення НЕ відсутнє.
# all(axis=1) перевіряє умову по всім колонкам в кожному рядку.
# Якщо в усіх питаннях рядка є відповіді, результат буде True.
complete_answers_mask = answers_to_schema_questions.notna().all(axis=1)

# sum() рахує кількість True (тобто 1).
complete_respondents = complete_answers_mask.sum()

print('Кількість респондентів, які відповіли на всі знайдені питання:', complete_respondents)
print('Відсоток таких респондентів, %:', round(complete_respondents / total_respondents * 100, 2))
```

```python
print('Аналітичний висновок до завдання №2:')
print(f'- Кількість респондентів без пропусків за всіма знайденими питаннями: {complete_respondents}.')
print(f'- Їх відсоток від загальної вибірки: {round(complete_respondents / total_respondents * 100, 2)}%.')
print('- Низьке або нульве, як в нашому випадку, значення цієї метрики є нормальним для великих опитувань: частина питань може бути не обовʼязковою, опціональною, залежати від попередніх відповідей або пропонуватися лише окремим групам респондентів.')
print('- Для подальшого аналізу неможна автоматично видаляти всі рядки з пропусками, т.я. це може призвести до значного скорочення вибірки та спотворювання висновків.')
```

# 3. Стастичний аналіз досвіду респондентів.

**Мета завдання:**
визначити міри центральної тенденції для поля WorkExp

Необхідно знайти:

- середнє значення (mean() - повертає середнє значення, проте чутливе до екстремально великих/малих значень);
- медіану (median() - повертає центральне значення і більш стійке до вибросів);
- моду (mode() - повертає значення, що зустрічається найчастіше).

Pandas за замовчуванням ігнорує пропуски в mean(), median() та mode(), проте, ми додатково перевіримо їх кількість.

```python
# Перевіримо наявність стовпця WorkExp.
'WorkExp' in df.columns
```

```python
# Переведемо стовпець WorkExp у числовий формат.
# Для безпечного переходу використаємо errors='coerce' - якщо значення неможливо конвертувати у число, тоді замінить його на NaN.

work_exp = pd.to_numeric(df['WorkExp'], errors='coerce')

work_exp_missing = work_exp.isna().sum()
work_exp_non_missing = work_exp.notna().sum()

print('Непустих значень WorkExp:', work_exp_non_missing)
print('Пропущених значень WorkExp:', work_exp_missing)
print('Відсоток пропусків WorkExp, %:', round(work_exp_missing / total_respondents * 100, 2))
```

```python
# Розрахуємо міри центральної тенденції.

work_exp = df['WorkExp'].dropna() # конкретно в цьому випадку, навіть якщо закоментувати цей пункт - результат не зміниться 

work_exp_mean = round(work_exp.mean(),2)
work_exp_median = work_exp.median()
work_exp_mode = work_exp.mode()

# mode() може повернути декілька значень, якщо ці значення зустрічаються однаково часто.
# щоб перевірити це, оформимо усі знайдені моди у список для виводу у таблиці.
work_exp_mode_list = work_exp_mode.tolist()

work_exp_stats = pd.DataFrame({
    'metric': ['mean', 'median', 'mode'],
    'value': [work_exp_mean, work_exp_median, work_exp_mode_list]
})

work_exp_stats
```

```python
print('Аналітичний висновок до завдання №3:')
print(f'- Середній досвід роботи: {round(work_exp_mean, 2)} років.')
print(f'- Медіанний досвід роботи: {round(work_exp_median, 2)} років.')
print(f'- Значення досвіда роботи, що зустрічається найчастіше: {work_exp_mode_list}.')
print('- Для характеристики типового респондента медіанне значення надійніше за середнє, т.я. воно в меншій мірі залежить від викидів.')
```

**Висновок:**  
В результаті маємо **значення середнього вище за медіану (13.37>10)**, це свідчить про те, що в датасеті є група спеціалістів з дуже великим періодом 'WorkExp', які "піднімають" середнє значення досвіду по датасету.  
**Медіана та мода збігаються (10=[10])**, тобто типовий спеціаліст у вибірці має досвід роботи близько 10 років, звідси можна зробити висновок, що опитування проходили досвідчені спеціалісти.

# 4. Аналіз віддаленої роботи.

**Мета завдання:**
визначити кількість респондентів, які працюють віддалено

```python
# Перевіримо наявність стовпця RemoteWork.
'RemoteWork' in df.columns
```

```python
# Дивимось розподіл значень в стовпці RemoteWork.
# value_counts(dropna=False) повертає і реальні категорії, і пропуски.
remote_work_distribution = df['RemoteWork'].value_counts(dropna=False)
remote_work_distribution
```

```python
# Розрахунок строго fully remote:

fully_remote_workers = df[df['RemoteWork'] == 'Remote']

fully_remote_count = fully_remote_workers.shape[0]
fully_remote_percent = round(fully_remote_count / total_respondents * 100, 2)
print('Кількість респондентів, які працюють віддалено:', fully_remote_count)
print('Відсоток від всіх респондентів, %:', fully_remote_percent)
```

```python
# Це додатковий розрахунок, через широку фільтрацію він включає часткове співпадіння за критерієм remote.
# Створюємо булеву маску для респондентів, які працюють віддалено.
# astype(str) переводить значення в рядки.
# str.contains() шукає слово remote всередині значення.
# case=False робить пошут нечутливим до регістру.
# na=False захищає код від пропусків.
remote_mask = df['RemoteWork'].astype(str).str.contains('remote', case=False, na=False)

remote_count = remote_mask.sum()
remote_percent = remote_count / total_respondents * 100
dif = round(remote_percent - fully_remote_percent, 2)

print('Кількість респондентів, які працюють віддалено та гібрідно:', remote_count)
print('Відсоток від всіх респондентів, %:', round(remote_percent, 2))
print('Різниця hybrid-remote, %:', dif)
```

```python
print('Аналітичний висновок до завдання №4:')
print(f'- Віддалено працюють {fully_remote_count} респондентів, що складає {fully_remote_percent}% вибірки.')
print('- Ця метрика допомагає оцінити розповсюдженність remote-формата серед розробників.')
```

**Висновок:**  
Датасет демонструє наявність декількох моделей роботи: fully remote, hybrid та flexisble office.  
Remote-формат роботи став повноцінною частиною IT-індустрії.  
Згідно даних датасету в форматі fully-remote працює більше 20% опитуваних, ще 13.69% працюють гібридно (частково віддалено), проте необхідно зазначити, що багато респондентів не дали відповідь на це запитання.  
Для ринку праці високий відсоток віддаленої роботи може означати велику конкуренцію між країнами і більш широкий вибір вакансій для спеціалістів.

# 7. Географічний аналіз компенсації Python-розробників.

**Мета завдання:**
серед респондентів, які використовують Python, розрахувати середню і медіанну річну компенсацію за країнами

```python
# Перевіримо наявність необхідних стовпців.
# Для корректного аналізу використовуємо вже конвертований в єдину валюту стовпець 'ConvertedCompYearly'.
required_columns_task_7 = ['LanguageHaveWorkedWith', 'Country', 'ConvertedCompYearly']
[col in df.columns for col in required_columns_task_7]
```

```python
# Створимо окрему таблицю тільки з Python-розробниками.
python_developers = df[python_mask].copy()

# Переведемо компенсацію у числовий формат на випадок, якщо наявні текстові значення.
python_developers['ConvertedCompYearly'] = pd.to_numeric(python_developers['ConvertedCompYearly'], errors='coerce')

# Приберемо рядки без країни та без компенсації.
python_compensation_clean = python_developers.dropna(subset=['Country', 'ConvertedCompYearly'])

print('Python-розробників всього:', python_developers.shape[0])
print('Python-розробників із вказаними країною та компенсацією:', python_compensation_clean.shape[0])
```

```python
# Згрупуємо дані за країною.
# agg() дозволяє порахувати за один раз декілька агрегованих метрик.
python_comp_by_country = (
    python_compensation_clean
    .groupby('Country')['ConvertedCompYearly']
    .agg(
        respondents_count='count',
        mean_compensation='mean',
        median_compensation='median'
    )
    .reset_index()
)

# Відсортуємо по медіанній компенсації, т.я. вона більш стійка до вибросів.
python_comp_by_country = python_comp_by_country.sort_values('median_compensation', ascending=False)

python_comp_by_country.head(20)
```

# 9. Аналіз популярності Python по віковим категоріям.

**Мета завдання:**
для кожної вікової категорії  визначити відсоток респондентів, які програмують на Python

```python
# Перевіряємо наявність стовпця Age.
'Age' in df.columns
```

```python
# Створимо таблицю для аналізу Python за віком.
age_python = pd.DataFrame({
    'Age': df['Age'],
    'uses_python': python_mask
})

# Приберемо рядки, в яких не вказаний вік.
age_python_clean = age_python.dropna(subset=['Age'])

# Групуємо за віком та розраховуємо метрики.
python_by_age = (
    age_python_clean
    .groupby('Age')['uses_python']
    .agg(
        respondents_count='count',
        python_users='sum'
    ).reset_index())

# Т.я. uses_python — булеве значення, sum() рахує кількість True.
python_by_age['python_percent'] = python_by_age['python_users'] / python_by_age['respondents_count'] * 100

# Відсортуємо за відсотком Python-розробників за спаданням.
python_by_age = python_by_age.sort_values('python_percent', ascending=False)

python_by_age
```

# 10. Аналіз індустрій серед високооплачуваних віддалених працівників.

**Мета завдання:**
знайти респондентів, які одночасно:
1. знаходяться вище 75-го перцентиля компенсації;
2. працюють віддалено;
3. мають вказану індустрію.  
Визначити найрозповсюдженіші індустрії в цій групі

```python
# Перевіримо наявність необхідних колонок.
required_columns_task_10 = ['ConvertedCompYearly', 'RemoteWork', 'Industry']
[col in df.columns for col in required_columns_task_10]
```

```python
# Створимо таблицю для аналізу та переведемо стовпець компенсації у числовий формат.
high_paid_remote = df[['ConvertedCompYearly', 'RemoteWork', 'Industry']].copy()
high_paid_remote['ConvertedCompYearly'] = pd.to_numeric(high_paid_remote['ConvertedCompYearly'], errors='coerce')

# Розрахуємо 75-й перцентиль тільки по непустим значенням компенсації.
# 75-й перцентиль визначає поріг, вище якого знаходиться верхня чверть значень по компенсації.
compensation_75_percentile = high_paid_remote['ConvertedCompYearly'].quantile(0.75)

print('75-й перцентиль річної компенсації:', compensation_75_percentile)
```

```python
# Створимо декілька умов фільтрації.
# 1. Компенсація вище або дорівнює 75-му перцентилю.
condition_high_comp = high_paid_remote['ConvertedCompYearly'] >= compensation_75_percentile

# 2. Формат роботи - строго fully remote.
condition_remote = (high_paid_remote['RemoteWork'] == 'Remote')

# 3. Індустрія вказана респондентом.
condition_industry_not_missing = high_paid_remote['Industry'].notna()

# Комбінуємо умови за домопогою &.
high_paid_remote_industry = high_paid_remote[condition_high_comp & condition_remote & condition_industry_not_missing]

print('Кількість високооплачуваних віддалених робітників із вказаною індустрією:', high_paid_remote_industry.shape[0])
```

```python
# Рахуємо найпоширеніші індустрії.
# value_counts() сортує значення за частотою від більшого до меншого.
top_industries_high_paid_remote = (
    high_paid_remote_industry['Industry']
    .value_counts()
    .reset_index())

top_industries_high_paid_remote.columns = ['Industry', 'respondents_count']

top_industries_high_paid_remote.head(20)
```

```python
print('Аналітичний висновок до завдання №10:')
print('- Визначено респондентів із високою компенсацією при fully remote-форматі роботи та вказаною індустрією.')
print('- Результат демонструє, в яких індустріях найчастіше зустрічаються високооплачувані спеціалісти, які працюють віддалено.')
```

**Висновок:**  
Посилаючись на дані датасету, **Software Development** — абсолютний лідер серед усіх індустрій для високооплачуваних спеціалістів, що працюють віддалено.  
Проте, для фінальних висновків важливо враховувати не тільки кількість, але і структуру всієї вибірки: великі індустрії можуть частіше попадати в топ через більше число респондентів.
